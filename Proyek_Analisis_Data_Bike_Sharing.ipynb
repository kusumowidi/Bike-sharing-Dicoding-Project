### Cleaning Data

```python
del hr_df
```

```python

drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

```python
# Mengubah nama judul kolom
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
# Mengubah angka menjadi keterangan
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Mengubah tipe data ke datetime dan categorical
day_df['dateday'] = pd.to_datetime(day_df.dateday)
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')
day_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
correlation_matrix = day_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(
    correlation_matrix,
    annot=True,
    mask=mask,
    cmap="coolwarm",
    center=0,
    fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

Heatmap korelasi digunakan untuk menunjukkan seberapa kuat hubungan antara setiap pasang variabel numerik dalam dataset. Semakin dekat nilai korelasi ke 1 atau -1, semakin kuat hubungan antara dua variabel tersebut. Nilai korelasi positif menunjukkan hubungan yang searah, sedangkan nilai korelasi negatif menunjukkan hubungan yang berlawanan arah. Nilai korelasi mendekati 0 menunjukkan bahwa dua variabel tersebut tidak berkorelasi.

Dalam konteks dataset ini, heatmap korelasi digunakan untuk mengeksplorasi hubungan antara variabel-variabel numerik seperti 'temp' (suhu), 'atemp' (feeling temperature), 'hum' (kelembaban), 'casual', 'registered', dan 'count' (jumlah total peminjaman sepeda).

Hasil analisis heatmap korelasi pada dataset ini menunjukkan beberapa temuan penting:

1. Variabel 'temp' dan 'atemp' memiliki korelasi positif yang sangat kuat, yang bermakna bahwa suhu aktual dan feeling temperature bergerak beriringan.
2. Variabel 'count' (jumlah total peminjaman sepeda) memiliki korelasi yang cukup kuat dengan 'casual' (jumlah peminjam casual) dan 'registered' (jumlah peminjam terdaftar). Hal ini bermakna bahwa keduanya berkontribusi terhadap jumlah total peminjaman sepeda.
3. Variabel 'hum' (kelembaban) memiliki korelasi negatif yang lemah dengan 'count', menunjukkan bahwa kelembaban yang lebih tinggi sedikit terkait dengan jumlah peminjaman sepeda yang lebih rendah.

Dengan demikian, heatmap korelasi memberikan pemahaman yang berguna tentang hubungan antara variabel-variabel numerik dalam dataset, yang dapat membantu dalam analisis lebih lanjut dan pengambilan keputusan.


## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perbedaan Pola penggunaan sepeda antara hari kerja dan hari libur?

```python
# 1. Bagaimana perbedaan Pola penggunaan sepeda antara hari kerja dan hari libur?

plt.figure(figsize=(10, 6))
sns.boxplot(data=day_df, x='workingday', y='count', hue='holiday')
plt.title('Bike Rentals: Working Day vs Holiday')
plt.xlabel('Working Day (1: Yes, 0: No)')
plt.ylabel('Total Bike Rentals')
plt.xticks(ticks=[0, 1], labels=['Non-Working Day', 'Working Day'])
plt.legend(title='Holiday', labels=['No', 'Yes'])
plt.show()
```

### Pertanyaan 2: Apa Korelasi antara kondisi cuaca dan jumlah peminjaman sepeda?

```python
# 2. Apa Korelasi antara kondisi cuaca dan jumlah peminjaman sepeda?
plt.figure(figsize=(10, 6))
sns.boxplot(data=day_df, x='weather_cond', y='count')
plt.title('Effect of Weather on Bike Rentals')
plt.xlabel('Weather Condition')
plt.ylabel('Total Bike Rentals')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 3: Bagaimana Perbandingan jumlah peminjaman sepeda antara musim panas dan musim dingin?

```python
# 3. Bagaimana Perbandingan jumlah peminjaman sepeda antara musim panas dan musim dingin?

plt.figure(figsize=(10, 6))
sns.boxplot(data=day_df, x='season', y='count')
plt.title('Comparison of Bike Rentals: All Seasons')
plt.xlabel('Season')
plt.ylabel('Total Bike Rentals')
plt.show()
```

### Pertanyaan 4: Bagaimana Tren harian jumlah peminjaman sepeda selama dua tahun?

```python
# 4. Bagaimana Tren harian jumlah peminjaman sepeda selama dua tahun?

plt.figure(figsize=(12, 6))
sns.lineplot(data=day_df, x='dateday', y='count')
plt.title('Daily Bike Rental Counts Over Two Years')
plt.xlabel('Date')
plt.ylabel('Count of Bike Rentals')
plt.show()
```
